```python
import pandas as pd
df = pd.read_csv('data/CPD_ParkArt.csv')
df.dtypes
```

```python
df["PARK"].nunique()
df["ARTIST"].nunique()
df["OWNER"].nunique()

df["PARK"].value_counts().head(15) # parks with how many art?
df["ARTIST"].value_counts().head(15) # art attributed how many times to which artist 
df["OWNER"].value_counts() # owner owns how many art pieces NOTE: CPD 231
```

```python
# df[["X_COORD", "Y_COORD"]].isna().sum()
print(df.isnull().sum())
```

```python
import plotly.express as px

fig = px.scatter_mapbox(
    df,
    lat="lat",
    lon="lon",
    hover_name="title",
    hover_data={
        "artist": True,
        "park_name": True,
        "year_installed": True,
        "lat": False,
        "lon": False
    },
    color="medium",   # or "artist"
    zoom=10,
    height=700
)

fig.update_layout(
    mapbox_style="open-street-map",
    mapbox_center={"lat": 41.8781, "lon": -87.6298},
    margin={"r":0, "t":0, "l":0, "b":0}
)

fig.show()
```
